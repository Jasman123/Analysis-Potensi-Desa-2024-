```python
import pdfplumber
import pytesseract
from PIL import Image
import pandas as pd
import os

file_path = "statistik-potensi-desa-indonesia-2024.pdf"

# Folder for output CSVs
output_folder = "desa_output_csv"
os.makedirs(output_folder, exist_ok=True)

# Dictionary: category name -> page numbers (1-indexed)
dict_table = {
    "KetenagaKerjaan": [51, 52, 53, 54],
    "Mata Air": [69],
    "Pendidikan": [80, 81, 82],
    "Kesehatan": [84, 85],
    "Transportation": [108, 109],
    "Industri": [136, 137, 138],
    "Crime": [215, 216, 217]
}

with pdfplumber.open(file_path) as pdf:
    for category, pages in dict_table.items():
        extracted_data = []  # list of dicts to convert to DataFrame later

        print(f"Processing category: {category} ({len(pages)} pages)")

        for page_number in pages:
            page_index = page_number - 1  # adjust to 0-based
            page = pdf.pages[page_index]

            # Convert to image (for OCR)
            im = page.to_image(resolution=300).original

            # OCR extraction
            text = pytesseract.image_to_string(im)

            # Save as structured record
            extracted_data.append({
                "Page": page_number,
                "Extracted_Text": text.strip()
            })

        # Convert to DataFrame
        df = pd.DataFrame(extracted_data)

        # Save as CSV using the category name
        output_path = os.path.join(output_folder, f"{category}.csv")
        df.to_csv(output_path, index=False, encoding="utf-8-sig")

        print(f"‚úÖ Saved: {output_path}")

print("\nüéâ All categories processed successfully!")
```

```python
import pdfplumber
import pytesseract
from PIL import Image
import re

file_path = "statistik-potensi-desa-indonesia-2024.pdf"

with pdfplumber.open(file_path) as pdf:
    # Example: human page 51 ‚Üí index 50
    page = pdf.pages[50]
    im = page.to_image(resolution=300).original
    text = pytesseract.image_to_string(im)

    # --- Extract TITLE (from start until "Aceh") ---
    title_match = re.search(r"^(.*?Aceh)", text, re.IGNORECASE | re.DOTALL)
    title_text = title_match.group(1).strip() if title_match else ""

    # --- Extract CONTENT (from "Aceh" until "Indonesia") ---
    content_match = re.search(r"(Aceh[\s\S]*?Indonesia)", text, re.IGNORECASE)
    content_text = content_match.group(1).strip() if content_match else ""

    # --- Print results ---
    if title_text:
        print(title_text)
    else:
        print("‚ùå No title section found (before 'Aceh').")

    print("\n" + "="*60 + "\n")

    if content_text:
        print(content_text)
    else:
        print("‚ùå No text found between 'Aceh' and 'Indonesia'.")
```

```python
import requests
import tabula
import pdfplumber
import pandas as pd
import camelot

url = "https://github.com/Jasman123/Analysis-Potensi-Desa-2024-/raw/main/statistik-potensi-desa-indonesia-2024.pdf"
file_path = "statistik-potensi-desa-indonesia-2024.pdf"

response = requests.get(url)
with open(file_path, "wb") as f:
    f.write(response.content)

print("✅ PDF downloaded successfully!")
```

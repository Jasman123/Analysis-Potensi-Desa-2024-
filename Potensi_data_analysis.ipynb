```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url_ket = "https://raw.githubusercontent.com/Jasman123/Analysis-Potensi-Desa-2024-/main/ketenagaKerjaan.xlsx"
url_pend = "https://raw.githubusercontent.com/Jasman123/Analysis-Potensi-Desa-2024-/main/pendidikan.xlsx"
url_ind = "https://raw.githubusercontent.com/Jasman123/Analysis-Potensi-Desa-2024-/main/Industri.xlsx"

df1 = pd.read_excel(url_ket)
df2 = pd.read_excel(url_pend)
df3 = pd.read_excel(url_ind)

def clean_dataframe(df):
  df = df.fillna(0)
  df = df.replace('-', 0)
  df = df.replace('=', 0)

  for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='ignore')

  return df

df1 = clean_dataframe(df1)
df2 = clean_dataframe(df2)
df3 = clean_dataframe(df3)
```

```python
#Analisi sumber daya dan penghasilan desa
df1.head()
```

```python
df1.describe()
```

```python
df1['Total Sumber Daya'] = df1.select_dtypes(include=np.number).sum(axis=1)

top_sda = df1.sort_values(by='Total Sumber Daya', ascending=False).head(10)

# Plot
sns.set_theme(style="whitegrid")  # clean background
plt.figure(figsize=(12,7))

# Create barplot
bar = sns.barplot(
    data=top_sda,
    x='Provinsi',
    y='Total Sumber Daya',
    palette='coolwarm',  # nicer professional palette
    edgecolor='black'
)

# Add value labels on top of bars
for p in bar.patches:
    height = p.get_height()
    bar.annotate(f'{int(height):,}',  # format with commas
                 (p.get_x() + p.get_width() / 2., height),
                 ha='center', va='bottom',
                 fontsize=10,
                 color='black')

# Rotate x labels for readability
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)

# Titles
plt.title('Top 10 Provinces by Total Sumber Daya', fontsize=16, weight='bold')
plt.xlabel('Provinsi', fontsize=12)
plt.ylabel('Total Sumber Daya', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
low_sda = df1.sort_values(by='Total Sumber Daya', ascending=True).head(10)
# Plot
sns.set_theme(style="whitegrid")  # clean background
plt.figure(figsize=(12,7))

# Create barplot
bar = sns.barplot(
    data=low_sda,
    x='Provinsi',
    y='Total Sumber Daya',
    palette='coolwarm',  # nicer professional palette
    edgecolor='black'
)

# Add value labels on top of bars
for p in bar.patches:
    height = p.get_height()
    bar.annotate(f'{int(height):,}',  # format with commas
                 (p.get_x() + p.get_width() / 2., height),
                 ha='center', va='bottom',
                 fontsize=10,
                 color='black')

# Rotate x labels for readability
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(fontsize=11)

# Titles
plt.title('Buttom 10 Provinces by Total Sumber Daya', fontsize=16, weight='bold')
plt.xlabel('Provinsi', fontsize=12)
plt.ylabel('Total Sumber Daya', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
df1.loc['Total Perkategori'] = df1.drop('Total Sumber Daya', axis=1).select_dtypes(include=np.number).sum()
top_kategori = df1.loc['Total Perkategori']
top_5_kategori = top_kategori.sort_values(ascending=False).head(5)
top_5_kategori
```

```python
bar_data = top_5_kategori.reset_index()
bar_data.columns = ['Kategori', 'Total']

# Bar plot
plt.figure(figsize=(8,6))
sns.barplot(
    data=bar_data,
    x='Kategori',
    y='Total',
    palette='viridis'  # nice professional color map
)

# Add value labels on top of bars
for i, value in enumerate(bar_data['Total']):
    plt.text(i, value + 0.02*max(bar_data['Total']), str(value), ha='center', fontsize=10)

plt.title('Total per Kategori (Top 5)', fontsize=14, weight='bold')
plt.ylabel('Total')
plt.xlabel('Kategori')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
